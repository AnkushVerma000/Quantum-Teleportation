```python
 from qiskit import QuantumCircuit, Aer, execute
from time import sleep

# Create a quantum circuit with one qubit
qc = QuantumCircuit(1, 1)

# Define a function to run the circuit
def run_circuit(input_bit):
    # Set the input bit
    if input_bit == 1:
        qc.x(0)

    # Flip the input bit using a NOT gate
    qc.x(0)

    # Measure the qubit and store the result in a classical register
    qc.measure(0, 0)

    # Execute the circuit on a simulator
    backend = Aer.get_backend('qasm_simulator')
    job = execute(qc, backend, shots=1)
    result = job.result().get_counts()

    # Print the result
    output = int(list(result.keys())[0])
    print('Sender:', input_bit, '| Receiver:', output)

    # Reset the qubit for the next run
    qc.reset(qc.qubits)

    return output

# Run the circuit repeatedly with a 2-second delay
data = []
while True:
    input_str = input('Enter state of sender electron (or N to stop): ')
    if input_str == 'N':
        break
    input_bit = int(input_str)
    output_bit = run_circuit(input_bit)
    data.append((input_bit, output_bit))
    sleep(1)

# Print the data
print('Teleported Data:')
for input_bit, output_bit in data:
    print(f' Receiver: {output_bit}')
    
# Store the output bits in a list
output_bits = [str(output_bit) for input_bit, output_bit in data]

# Concatenate the output bits to form a binary string
output_str = ''.join(output_bits)

# Print the binary string
print('Receiver Data:', output_str)



# Define the binary string
binary_str = output_str


# Split the string into groups of 8 bits (1 byte)
bytes_list = [binary_str[i:i+8] for i in range(0, len(binary_str), 8)]

# Convert each byte to its corresponding ASCII character
text = ""
for byte in bytes_list:
    decimal_value = int(byte, 2)
    character = chr(decimal_value)
    text += character

# Print the resulting text
print('Message Send by Sender :',text)
```

```python
text = input("Enter the text to convert to binary: ")

binary = ''.join(format(ord(char), '08b') for char in text)

print("Binary code: ", binary)

text = binary
new_text = ""
for c in text:
    if c == '0':
        new_text += '1'
    elif c == '1':
        new_text += '0'
    else:
        new_text += c
print("Value for Sender Electron",new_text)


```
